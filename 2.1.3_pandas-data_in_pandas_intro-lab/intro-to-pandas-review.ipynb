#### 4. Rename any columns with spaces or special characters to not have them

```python
# A:
import string
def clean_string(name):
    letter=string.punctuation
    space=string.whitespace
    spcl_char=letter+space
    pure_name=''
    for ch in name:
        if ch not in spcl_char:
            pure_name+=ch
    return pure_name
```

```python
cast_new_columns=[clean_string(col) for col in cast_columns]
cast_new_columns
```
