````python
#hypotesis
alpha=0.95
# calculate point estimate
sample_mean= np.mean(alchol)
# calculate test statastics
t_statastics= (np.mean(alchol)-np.median(alchol))/scipy.stats.sem(alchol)
print "t_statastics:", t_statastics
# find p_value
p_value=t.sf(np.abs(t_statastics), len(alchol)) * 2 
# two sided test
if p_value < alpha:
    print "We reject our null hypothesis and conclude that the true mean alchol value is different from the median alchol value."
elif p_value > alpha:
    print "We fail to reject our null hypothesis and cannot conclude that the true mean alchol value is different from the median ."
else:
    print "Our test is inconclusive."
````

````python
def percentile_based_outlier(data, threshold=95):
    diff = (100 - threshold) / 2.0
    minval, maxval = np.percentile(data, [diff, 100 - diff])
    return (data < minval) | (data > maxval)
````

````python
percentile_based_outlier(sat_out, threshold=95)
````

````python
def reject_outliers(sat_out, m=1):
    return sat_out[abs(sat_out - np.mean(sat_out)) < m * np.std(sat_out)]
````

````python
k=reject_outliers(sat_out, m=1)
percentile_based_outlier(k, threshold=95)
````

#How is percentile related to the spearman rank correlation?
A challenge in measuring correlation is that the variables we want to compare might not be expressed in the same units.
For example, height might be in centimeters and weight in kilograms. And even if they are in the same units, they come
from different distributions.

There are two common solutions to these problems:

Transform all values to standard scores. This leads to the Pearson coefficient of correlation.
Transform all values to their percentile ranks. This leads to the Spearman coefficient.
If X is a series of values, xi, we can convert to standard scores by subtracting the mean and dividing by the standard 
deviation: zi = (xi  ) / .
The numerator is a deviation: the distance from the mean. Dividing by  normalizes the deviation, so the values of Z 
are dimensionless (no units) and their distribution has mean 0 and variance 1.

If X is normally-distributed, so is Z; but if X is skewed or has outliers, so does Z. In those cases it is more robust
to use percentile ranks. If Rcontains the percentile ranks of the values in X, the distribution of Ris uniform between
0 and 100, regardless of the distribution of X.
