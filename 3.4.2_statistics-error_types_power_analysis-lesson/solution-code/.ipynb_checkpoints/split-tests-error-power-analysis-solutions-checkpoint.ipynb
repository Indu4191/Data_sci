### Introduction to A/B testing

---

You may have heard the term "A/B Testing" or "Split Testing" before. A split test is simply a business term for conducting an experiment testing different versions of your product with users. At the end of the test, you as the data scientist will statistically analyze the experiment and determine a "winner' according to a pre-defined metric. 

**Example: selling dog collars**

You work for a start-up that sells dog collars. Your web development team has constructed a prototype for a new "landing page" on the website (the landing page is just the first page that users reach when coming to the website). The designers are not sure whether or not having a picture of a black lab or a golden retriever wearing the collar will have an impact on click-through rate (the proportion of users that continue from the landing page to the the rest of the website). 

The company decides to run an A/B test to quantitatively evaluate which picture to choose:
- **Arm A** is the version of the landing page with the black lab
- **Arm B** is the version of the landing page with the golden retriever

For two weeks users will be directed at random with equal probability to one of the two landing pages. At the conclusion of two weeks, the click-through rates of each arm will be compared and one of the two arms will be "winnered".

Desiging and evaluating A/B tests like this one is one of the most common tasks that a data scientist will need to do at a company.

**Below are the click-through rates per arm at the end of two weeks.**

```python
A_wins = 113
A_loss = 87
B_wins = 87
B_loss = 103
```

<a id='ab-hypothesis'></a>

### Split tests are hypothesis tests

---

Despite the business jargon, **split tests are just experiments to test hypotheses.** Using the scenario above, we can frame the null hypothesis:

> **H0:** the difference in click-through rates between arms is 0.

And the alternative hypothesis:

> **H1:** the difference in click-through rates between arms is not 0.

It is very important that the users sent into arms are selected at random. If there something impacts which arm users are assigned to, such as whether they are viewing the site on web or mobile, then the arms have **selection bias**.

**What is the problem with making a decision about which dog picture to choose if users were not selected randomly to be in each arm?**

### The $\chi^2$ (chi-squared) test of independence

---

A popular Frequentist test to evaluate A/B tests is the $\chi^2$ test of independence. The $\chi^2$ test of independence is appropriate when you have categorical data and want to evaluate whether two groups are significantly different. 

Click-through rate can be thought of as binary categorical data: a user either clicked-through (1) or not (0). 

The "independence" refers to whether the outcome for the groups (the click-through rate) is independent of which group they were in. Independence would mean that there is no relationship between the dog picture and the click-through rate. 

You can conduct the $\chi^2$ test manually using what is known as a contingency table. [For a detailed overview of the procedure this site is a good resource.](https://onlinecourses.science.psu.edu/stat500/node/56) We will use python instead of manually calculating the test. That being said, it is important to address the formula for the $\chi^2$ statistic.

**The $\chi^2$ statistic**

### $$ \chi^2 = \sum_{i=1}^{cells} \frac{(O_i - E_i)^2}{E_i} $$

Where $cells$ refers to the number of cells in the contingency table, $O$ are the observed values (frequencies) and $E$ are the *expected* frequencies under perfect independence. 

**Using `stats.chi2_contingency` calculate the $\chi^2$ statistic and the associated p-value for our split test.**

```python
table = np.array([[A_wins, A_loss],
                  [B_wins, B_loss]])

results = stats.chi2_contingency(table)
chi2 = results[0]
pvalue = results[1]
print chi2, pvalue
```

**Explain what the p-value means in the context of our split test.**

```python
# We have a p-value of about 0.044. This means that there is a 4.4% chance that we 
# observed the difference in click-through rates between arms by chance due to 
# randomness in our sample of data when in fact there is no difference in click-through
# rates between arms in the overall population (if everyone in the world were to 
# visit our site).
```

**Which arm is the "winner", should you choose to accept it as such? By how much?**

```python
# Arm A has the higher click-through rate and would be the winner if this level
# of significance were sufficient.

A_rate = float(A_wins)/(A_wins+A_loss)
B_rate = float(B_wins)/(B_wins+B_loss)
A_vs_B = A_rate - B_rate
print A_rate, B_rate
print A_vs_B
print A_rate/B_rate
```

### Statistical significance, p-values, alpha threshold, and Type I error

---

The split test has concluded and you have performed your statistical analyisis of the results:
- **Arm A (black lab picture) has a 23% higher click-through rate than Arm B (golden retriever)**
- **The p-value of our test was 0.044**

Should we accept that the black lab picture is in fact better? Do we believe that the difference is real?

**Statistical significance**

It is common to see Frequentist tests reported as "statistically significant with p < 0.05" or with p < 0.01. What does it mean for a test to be statistically significant? On the surface, these statements are simply saying that the calculated p-value was less than a specific value. The values of 0.05 and 0.01 are common in academic research, but also arbitrary. 

What does having "p < 0.05" specifically mean?

> **p < 0.05**: In hypothetical repetitions of this experiment with the same sample size, fewer than 5% of the experiments would have measured a difference between arms at least this extreme by chance.

Likewise for p < 0.01, but instead framed in the context of the null and alternative hypothesis:

> **p < 0.01**: There is less than a 1% chance of accepting the alternative hypothesis when in fact the null hypothesis was true. 

---

**Type I error and the alpha threshold $\alpha$**

As rigorous researchers we would set a threshold for how likely we are to falsely accept the alternative hypothesis prior to running our experiment. This chance of incorrectly accepting the alternative hypothesis is known as **Type I error**.

Type I error is directly related to the p-value: 

> **A p-value is the risk of Type I error given the sample size and measured effect.**

It is important to set a threshold for Type I error before the experiment begins. This prevents us from arbitrarily deciding after we see the p-value whether we will accept or not.

The threshold we set for the Type I error rate is denoted with alpha. For example:
- $\alpha$ = 0.05 corresponds to a p < 0.05 significance level.
- $\alpha$ = 0.01 corresponds to a p < 0.01 significance level.

Furthermore, alpha is equivalent to 1. - confidence. A 95% confidence interval, for example, is associated with a Type I error threshold of $\alpha$ = 0.05.

**Aside on p-value thresholds:** 

p < 0.01 is historically considered a "conservative" significance threshold. In fact, it's not very conservative at all. 

This is at worst a 1/100 chance that a result is accepted when in fact the result is null. Think about all the papers written and experiments run that that have used a threshold of p < 0.01 to validate their findings... I guarantee there are a whole lot more than 100 of them.


<a id='type2-power'></a>

### Type II error ( $\beta$ )  and Power ( $1 - \beta$ )

---

We tend to focus a lot on Type I error, or the probability of false positives. It's counterpart, Type II error, is no less important. Type II error measures the risk of false negatives, or accepting the null hypothesis when in fact the alternative hypothesis is true.

**Alpha and beta**

Whereas alpha is the probability that we could have gotten our effect by chance due to sampling error, it's Type II counterpart beta is the probability that we accepted the null hypothesis by chance when in fact the alternative hypothesis is true.

**Statistical power**

It is rare to see beta. More commonly you will see $1 - \beta$, which is known as the power of the test. Think of the  power of a test as the probability of correctly accepting the alternative hypothesis when it is true.

<a id='power-visual'></a>

### Visualizing the interaction of alpha and power

---

Below is an interactive plot that helps visualize the relationship between the alpha threshold and the power of the experiment. 

- Decreasing alpha (making the threshold more strict) necessarily reduces our power. 
- Alternatively, the more lenient we are about Type I error rate the more we reduce Type II error rate.

```python
from ipywidgets import *
from IPython.display import display
```

```python
def power_plotter(std, alpha, effect):
    unit_norm = stats.norm(0,1)
    z_alpha = unit_norm.ppf(1-alpha/2.)
    z_power = (effect/std) - z_alpha 
    
    null_dist = stats.norm(0, std)
    alt_dist = stats.norm(effect, std)
    
    power = unit_norm.cdf(z_power)
    
    z_type1 = null_dist.ppf(1-alpha/2)
    
    xvals = np.linspace(-5*std, effect+(5*std), 300)

    fig, axarr = plt.subplots(2, 1, figsize=(14,6))

    axarr[0].plot(xvals, null_dist.pdf(xvals), color='grey', lw=4, label='Null')
    axarr[1].plot(xvals, alt_dist.pdf(xvals), color='black', lw=4, label='Alternative')

    type1_x = np.linspace(z_type1, effect+(5*std))

    axarr[0].axvline(z_type1, lw=3, ls='dashed', color='darkred')
    axarr[0].fill_between(type1_x, 0, null_dist.pdf(type1_x), color='darkred', 
                          alpha=0.5, hatch = '///', label='alpha = {:.2f}'.format(alpha))
    
    axarr[1].axvline(z_type1, lw=3, ls='dashed', color='steelblue')
    axarr[1].fill_between(type1_x, 0, alt_dist.pdf(type1_x), color='steelblue', 
                          alpha=0.5, hatch = '///', label='power = {:.2f}'.format(power))


    axarr[0].set_ylim([0, 0.50])
    axarr[1].set_ylim([0, 0.50])
    
    axarr[0].set_xlim([-5,10])
    axarr[1].set_xlim([-5,10])
    
    axarr[0].legend(loc='upper left', fontsize=14)
    axarr[1].legend(loc='upper left', fontsize=14)

    plt.show()
    
```

```python
def power_calcplotter(std_effect=1.0, alpha=0.05, effect=1.0):
    power_plotter(std_effect, alpha, effect)
    

interact(power_calcplotter, 
         std_effect=FloatSlider(min=0.5, max=3.0, value=1.0, step=0.1),
         alpha=FloatSlider(min=0.01, max=0.50, value=0.05, step=0.01),
         effect=FloatSlider(min=0.1, max=5.0, value=1.0, step=0.1))
```

### Power analysis and sample size calculation

---

We are now ready to formalize the relationship between the components of our experimental design and the data that we measure in the experiment.

In this context we are not analyzing data already gathered, but instead planning an experiment to test a hypothesis. 

There are five components we consider when setting up our experiment. 
1. Our desired Type I error rate ($\alpha$)
2. Our desired Type II error rate ($\beta$ or more commonly power, $1 - \beta$)
3. The expected size of the effect, or mean difference between groups
4. The expected standard deviation of measurement
5. The sample size

Often the expected effect size and the standard deviation of measurement will need to be estimated from prior work or other information.

---

We can formulate an equation to solve for any one of these components given the others. Below are the calculations for the power of a test and the calculation for the required sample size of a test.

Given variables:
- $\theta$ : a test statistic measuring the difference between groups
- $SE_{\theta}$ : the standard error of measurement of the test statistic
- $\alpha$ : the pre-specified Type I error threshold value
- $Z_{\alpha/2}$ : the 2-tailed z-statistic for the specified $\alpha$ on the unit Normal distribution
- $n_1$ : the sample size of group 1
- $n_2$ : the sample size of group 2

**Calculation for statistical power of a test:**

The z-statistic for the power, $1 - \beta$, is calculated:

### $$ Z_{power} = \frac{\theta}{SE_{\theta}} - Z_{\alpha/2} $$

**Calculation for the required sample size of a test:**

The formula above can be re-written with $\sigma^2$, the (assumed) equal standard deviation of measurements in the two groups.

Let $r$ be the ratio of group 1 sample size to group 2 sample size

### $$ Z_{power} = \frac{\theta}{ \sqrt{ \frac{\sigma^2}{n_1} + \frac{\sigma^2}{rn_1} } } - Z_{\alpha/2} $$

Solve for $n_1$ in the previous equation.

### $$ n_1 = \frac{r+1}{r} \frac{\sigma^2 (Z_{power} + Z_{\alpha/2})^2}{\theta^2} $$

If you assume that the two groups will have equal sizes (which is most often the case when *preparing* a test: the best statistical power for the fewest number of subjects is achieved with equal groups), $r = 1$ and the formula reduces to:

### $$ n = \frac{2 \sigma^2 (Z_{power} + Z_{\alpha/2})^2}{\theta^2} $$

where $n$ is the number of subjects requred in each group.

**Note: Sample size calculations are hypothetical**

These equations are guidelines for the optimal setup of an experiment and should therefore be fairly conservative. That being said, there is good reason to not exceed the required number of subjects by too much: there are marginal returns when using more subjects, and tests are often costly to run.

<a id='visual-equation'></a>

### Calculating the required sample size visually

---

The plotting function below will plot the interacting components in the equations above to give you a visual feel for how the different parts interact.

As the user, you will need to supply to the function `required_sample_plotter`:
- `std`: the expected standard deviation of measurements
- `desired_power`: the statistical power threshold for the test
- `alpha`: the Type I error rate threshold for the test
- `effect`: the expected effect between groups (standardized to units of standard deviation of the measurement) 

The function will plot the distribution of measurements from the hypothetical groups, coloring in the parts of the distribution corresponding to alpha and power. The estimated number of subjects required is displyed in the legend of the plot.

```python
def n_calculator(std, desired_power, alpha, effect):
    unit_norm = stats.norm(0,1)
    z_power = unit_norm.ppf(desired_power)
    z_alpha = unit_norm.ppf(1-alpha/2.)
    n = np.ceil((2. * std * (z_power + z_alpha)**2) / (effect**2))
    return int(n)
```

```python
def required_sample_plotter(std=1, desired_power=0.8, alpha=0.05, effect=0.5):
    n = n_calculator(std, desired_power, alpha, effect)
    
    fig, ax = plt.subplots(figsize=(12,5))
    
    null_dist = stats.norm(0, std)
    alt_dist = stats.norm(effect, std)
        
    null_means = np.array([np.mean(null_dist.rvs(n)) - np.mean(null_dist.rvs(n)) for i in range(1000)])
    alt_means = np.array([np.mean(alt_dist.rvs(n)) - np.mean(null_dist.rvs(n)) for i in range(1000)])
    
    alpha_null_val = stats.scoreatpercentile(null_means, (1. - (alpha/2))*100)
    
    pct_alt_above = np.sum(alt_means > alpha_null_val)/1000.
    pct_null_above = np.sum(null_means > alpha_null_val)/1000.
        
    ax = sns.distplot(null_means[null_means < alpha_null_val],
                      ax=ax, color='#333333', kde=False, hist_kws=dict(alpha=0.5, lw=2, 
                                                                       histtype='step'))
    ax = sns.distplot(null_means[null_means >= alpha_null_val],
                      ax=ax, color='darkred', kde=False, hist_kws=dict(alpha=0.6),
                      label='alpha')
    
    ax = sns.distplot(alt_means[alt_means >= alpha_null_val], ax=ax,
                      color='steelblue', kde=False, hist_kws=dict(alpha=0.2),
                      label='power (1 - beta)')
    ax = sns.distplot(alt_means[alt_means < alpha_null_val], 
                      ax=ax, color='#333333', kde=False, hist_kws=dict(alpha=0.5, lw=2,
                                                                       histtype='step'))
    
    ax.axvline(alpha_null_val, lw=3, ls='dashed', color='black')
    
    n_str = 'N = '+str(int(n))
    effect_str = 'effect size = {:.2f}'.format(float(effect)/std)
    power_str = 'power = {:.3f}'.format(pct_alt_above)
    alpha_str = 'alpha = {:.3f}'.format(pct_null_above)
    
    ax.annotate('\n'.join([n_str, effect_str, alpha_str, power_str]),
                xy=(0.95, 0.6),
                xycoords='axes fraction', horizontalalignment='right', 
                fontsize=14, weight='bold', color='black')
    
    ax.legend(loc='upper left')
    
```

```python
deltas = required_sample_plotter(std=1, desired_power=0.5, alpha=0.05, effect=0.2)
```
